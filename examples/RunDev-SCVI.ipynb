```python
# various import statements
import os
import inspect
import seaborn
import matplotlib
import matplotlib.pyplot as plt
import torch
import scanpy as sc
import scvi

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

if torch.cuda.is_available():
    print("GPU is available")
    print("Number of GPUs:", torch.cuda.device_count())
    print("GPU Name:", torch.cuda.get_device_name(0))
else:
    print("GPU is not available")

torch.set_float32_matmul_precision('medium')
```

```python
adata=sc.read_h5ad(os.path.expanduser('/allen/programs/celltypes/workgroups/rnaseqanalysis/EvoGen/Team/Matthew/data/taxtest/HvQvM/HvQvMall_cere_clean.h5ad'),backed='r')

batch_key='batch_name'
discov_key='species'
layer_key='spliced'
dataset_key='dataset_name'
```

```python
scvi.model.SCVI.setup_anndata(adata, layer=layer_key, batch_key=batch_key,categorical_covariate_keys=[discov_key,dataset_key])#'roi',,
scvi_model = scvi.model.SCVI (adata, n_latent=150,n_layers=3,n_hidden=512)
```

```python
scvi_model.train(100)
```

```python
SCVI_LATENT_KEY = "X_SCVI"
adata.obsm[SCVI_LATENT_KEY] = scvi_model.get_latent_representation(adata)
```

```python
MDE_KEY = "X_SCVI_MDE"
adata.obsm[MDE_KEY] = scvi.model.utils.mde(adata.obsm[SCVI_LATENT_KEY],init='random')

sc.pl.embedding(
    adata,
    basis=MDE_KEY,
    color=[discov_key,batch_key,'dataset_name'],palette=sc.pl.palettes.godsnot_102,cmap='coolwarm'
)

sc.pl.embedding(
    adata,
    basis=MDE_KEY,
    color=['region','general_region'],palette=sc.pl.palettes.godsnot_102,cmap='coolwarm'
)
```
